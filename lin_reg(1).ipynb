```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
money_wallet=(np.array([100,65,80,50,55]))
```

```python
donation=(np.array([10,5,7,3,6]))
```

```python
plt.xlabel('Money in wallet')
plt.ylabel('Donations')
plt.scatter(money_wallet,donation)
```

```python
plt.xlabel('Money in wallet')
plt.ylabel('Donations')
plt.scatter(money_wallet,donation)
plt.axhline(np.mean(donation), color='red')
```

```python
def calculate_sse(data_input,fit_val):
    sse=0
    squared_errors=[]
    for point in data_input:
        squared_error=(point-fit_val)**2
        sse+=squared_error
        squared_errors.append(int(squared_error))
    return sse,squared_errors
```

```python
sse,error_list=calculate_sse(donation,np.mean(donation))
```

```python
sse
```

```python
error_list
```

```python
plt.ylabel('Errors')
plt.xticks([0,1,2,3,4,5])
plt.plot(error_list)
```

```python
slope, intercept=np.polyfit(money_wallet,donation,deg=1)
```

```python
fig, ax=plt.subplots()
ax.scatter(money_wallet,donation)
ax.plot(money_wallet,slope*money_wallet+intercept,'-',color='red')
```
